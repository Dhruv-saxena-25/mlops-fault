```python
from scipy.stats import ks_2samp
import numpy as np
```

```python
np.random.seed(12345678)
```

```python
x = np.random.normal(0, 1, 1000)
```

```python
y = np.random.normal(0, 1, 1000)
```

```python
z = np.random.normal(1.1, 0.9, 1000)
```

```python
ks_2samp(x, y)
```

```python
ks_2samp(x, z)
```

####  If p-value is lower than a, then it is very probable that the two distributions are different.

```python
# -- the null hypothesis is that the samples are drawn from the same distribution.
```

```python
# A small p-value means we reject the null hypothesis.
```

```python
# So in the following example, p-value is very small, which means we reject the null hypothesis and conclude that the two samples x and z are not drawn from the same distribution.
```

```python
import seaborn as sns
import pandas as pd
sns.ecdfplot(pd.DataFrame({'x': x, 'z': z}))
```
